# Model
For the neural network I wrote up a class in python that takes in a list of integers which specify the size of each layer. For example, in my implementation I put 4 layers with the first being 784 nodes big, then 30, 30, and the output layer is only 10 nodes big.

The class is included in the source code if you'd like to look at it, but the important things to know are that I am implementing Stochastic Gradient Descent with backpropogation to fine tune our predictions.

```python
import network
import mnist_loader

training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
```

```python
net = network.Network([784, 30, 30, 10])

net.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```
